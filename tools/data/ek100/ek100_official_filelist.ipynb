```python
for id_domain in range(3):
    print(f'D{id_domain+1}')
    for split in ['train', 'test']:
        file_stem = f'D{id_domain+1}_{split}'
        df = pd.read_pickle(f'/data/gunsbrother/repos/CoMix/video_splits/{file_stem}.pkl')
        df['path'] = df['video_id']
        df['length'] = df['stop_frame'] - df['start_frame'] + 1
        df[['path', 'start_frame', 'length', 'verb_class']].to_csv(
            f'/data/gunsbrother/repos/haawron_mmaction2/data/_filelists/ek100/mm-sada_resampled/{file_stem}.txt',
            sep=' ', header=False, index=False)
        print(f'    {split:5s}: N = {len(df):4d}, T = {df["length"].sum()/3600:5.1f} h, K = {len(np.sort(df["verb_class"].unique()))}')
```
